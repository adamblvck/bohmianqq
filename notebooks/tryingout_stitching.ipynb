```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
size = 512
x = np.linspace(-20, 20, size)
y = np.linspace(-20, 20, size)
```

```python
g = 9.81
```

```python
linear_fall = g*x
```

```python
# My solution
c = -40
D = -5
a = c / D**2
b = g - 2*D*a

parabola = a*x**2 + b*x + c
```

```python
# Now plot the spacetime density
# plot_spacetime(prob_arr, z_array, t_vals, title="Boundary Absorption Test - Cosine Mask")
fig, ax = plt.subplots(1, 1, figsize=(11, 4))

# density plot
im = ax.plot(x, linear_fall)
im = ax.plot(x, parabola)

ax.set_xlabel("time Δt")
ax.set_ylabel("z")
ax.set_title("|ψ(z,t)|² with Polynomial CAP")
plt.tight_layout()
plt.show()
```
